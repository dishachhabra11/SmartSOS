```python
import os

# Define the path to the folder
folder_path = '/content/drive/MyDrive/sos/TESS'

paths=[]
labels=[]
# Walk through the folder and its subdirectories


# Iterate through subfolders in the root folder
for subdir in os.listdir(folder_path):
    subdir_path = os.path.join(folder_path, subdir)
    if os.path.isdir(subdir_path):
        # Iterate through files in each subfolder
        for file_name in os.listdir(subdir_path):
            # Print the name of the file
            label=file_name.split('_')[2];
            label=label.split('.')[0].lower()
            labels.append(label)

            path=os.path.join(subdir_path,file_name)
            paths.append(path)
print('data set loaded')






```

```python
def plot_waveform(signal, sr, emotion):
    # Plot the waveform
    plt.figure(figsize=(10, 4))
    librosa.display.waveshow(signal, sr=sr)
    plt.title('Waveform for Emotion: {}'.format(emotion))
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

def plot_stft(signal, sr, emotion):
    # Compute the STFT
    stft = librosa.stft(signal)

    # Convert to dB scale (log scale) for better visualization
    stft_db = librosa.amplitude_to_db(abs(stft))

    # Plot the STFT
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(stft_db, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('STFT for Emotion: {}'.format(emotion))
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()
```

```python
emotion='fear'
path=df['speech'][df['label']==emotion].iloc[0]

signal,sr=librosa.load(path)
plot_waveform(signal,sr,emotion)

plot_stft(signal,sr,emotion)
Audio(path)
```

feature extraction


```python
def extract_mfcc(audio_file_path, n_mfcc=40):
    # Load the audio file
    y, sr = librosa.load(audio_file_path,duration=3,offset=0.5)

    # Extract MFCC features
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc).T,axis=0)


    return mfccs
extract_mfcc('/content/drive/MyDrive/sos/TESS/OAF_Fear/OAF_bar_fear.wav')
```

```python
X_mfcc=df['speech'].apply(lambda x:extract_mfcc(x))
```

```python
X_mfcc
```

```python
X=[x for x in X_mfcc]
X=np.array(X)
X.shape
```

```python
X = np.expand_dims(X, axis=-1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()
y=enc.fit_transform(df[['label']])
y=y.toarray()
```

MODEL


```python

from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout

model=Sequential([
    LSTM(123, return_sequences=False, input_shape=(40, 1)),
    Dense(64, activation='relu'),
    Dropout(0.1),
    Dense(32, activation='relu'),
    Dropout(0.5),  # Corrected dropout rate
    Dense(7, activation='softmax')  # Co
])
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()

```

```python
history=model.fit(X,y,validation_split=0.2,epochs=100,batch_size=516,shuffle=True)
```
